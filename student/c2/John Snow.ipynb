## Question 1: Correlation vs. Causation
The map showed a high concentration of deaths near the Broad Street pump. However, proximity alone only proves a correlation. What specific steps did Snow take to move his argument from correlation (deaths happen near the pump) to causation (the pump water is causing the deaths)?

Snow went beyond simply observing the map. He investigated the people who died and looked for patterns between them. He also examined unusual cases people who lived near the pump but did not get sick—and found that they used a different water source. In addition, he identified people who lived far away but became ill because they drank water from the Broad Street pump. These observations helped him demonstrate that the water from the pump was the actual cause of the disease, not just a correlation.

## Question 2: The Danger of Missing Data
Imagine if Snow had only collected data from the local hospitals, rather than walking the streets and knocking on doors. How would this have changed his dataset, and how might it have altered his conclusions?

If Snow had only collected data from hospitals, the dataset would have been incomplete. Not all people go to hospitals, and some may have died at home. In addition, hospitals might not collect detailed information about patients’ lifestyles or water sources. This would result in missing important data, leading to an incomplete picture of the outbreak and possibly incorrect conclusions.
