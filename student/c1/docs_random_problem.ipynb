## 1. توليد كلمة مرور

اكتب إجراءً يقبل المعاملات التالية ليُنتِج **كلمة مرور** (Password):

- الطول
- الأحرف الصغيرة
- الأحرف الكبيرة
- الأرقام
- الحد الأدنى من الأرقام
- الرموز: `!@#$%^&*`
- الحد الأدنى من الرموز

```python
def generate_password(length: int, ...) -> str:
    ...
```

````python
import random
import string

def generate_password(length: int,
                      use_lower=True,
                      use_upper=True,
                      use_digits=True,
                      min_digits=1,
                      use_symbols=True,
                      min_symbols=1) -> str:
    
    lower = string.ascii_lowercase
    upper = string.ascii_uppercase
    digits = string.digits
    symbols = "!@#$%^&*"
    
    pool = ""
    
    if use_lower:
        pool += lower
    if use_upper:
        pool += upper
    if use_digits:
        pool += digits
    if use_symbols:
        pool += symbols
    
    password = []
    
    # نضمن الحد الأدنى من الأرقام
    for _ in range(min_digits):
        password.append(random.choice(digits))
    
    # نضمن الحد الأدنى من الرموز
    for _ in range(min_symbols):
        password.append(random.choice(symbols))
    
    # نكمل باقي الطول
    while len(password) < length:
        password.append(random.choice(pool))
    
    # نخلط الترتيب
    random.shuffle(password)
    
    return ''.join(password)

print(generate_password(10))
````
